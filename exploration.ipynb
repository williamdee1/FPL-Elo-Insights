# Linking Players and Matches:

```python
def player_info(player_id):
    """
    Returns a dictionary of information for a given player.
    """
    player = players_master[players_master["player_id"] == player_id]
    if player.empty:
        return None
    return player.iloc[0].to_dict()
```

```python
print(player_info(1))
id_1_player = player_info(1)
print(id_1_player['position'])  # Example usage to get position
```
